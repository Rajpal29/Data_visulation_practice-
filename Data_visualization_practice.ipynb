```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic graph
##### https://matplotlib.org/ 
##### https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html - helpful links


```python
# plt.plot([1,2,3], [2,4,6]) # to clean this up we should move our arrays into variables 

x = [1,2,3,4]
y = [2,4,6,8]

# Resize your graph
plt.figure(figsize=(5,3), dpi=150)



plt.plot(x,y, label="2x", color="orange", linewidth=2, linestyle="--", marker=".", markersize=10, markeredgecolor="black") # when we use .legned we need to use label="...."

# Shorthand notation
# fmt = "[color][marker][line]"
# plt.plot(x,y, "b^--", label="2x")

## Line number two
 
x2 = np.arange(0,4,0.5)
plt.plot(x2[:6], x2[:6]**2, "r", label="X^2")
plt.plot(x2[5:], x2[5:]**2, "r--")

plt.title("Our First Graph", fontdict={"fontname": "Comic Sans MS", "fontsize": 20}) # title, font, fontsize  
plt.xlabel("X Axis") # x axis
plt.ylabel("Y Axis") # y axis

# X and Y labels
plt.xticks([0,1,2,3,4]) # changing the ticks before on the x axis we had float number e.g. 1.0, 1.5 etc
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig("mygraph.png", dpi=300) # this would save in the current directory with a dpi of 300 so graph has high resolution

#shows plot
plt.show() # gets ride of [<matplotlib.lines.Line2D at 0x22c9d73da00>] above the graph
```

### Bar chart

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

# bars[0].set_hatch("/")
# bars[1].set_hatch("O")
# bars[2].set_hatch("*")

#another way of applying the patterns 

patterns = ["/", "O", "*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
plt.figure(figsize=(6,4))

plt.show()
```
